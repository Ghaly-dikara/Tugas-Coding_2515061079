###Vector Transpotition

```python
# lakukan transpose pada vector x dan simpan pada variabel x_t
x = np.array([[0.79,7,9]])
x_t = x.T
x_t
```

```python
# tampilkan ukuran dari vector x_t
x_t.shape
```

```python
# lakukan transpose pada vector y dan simpan pada variabel y_t
y = np.array([0,7,9])
y
```

```python
y.shape
```

```python
# lakukan transpose pada vector y dan simpan pada variabel y_t
y = np.array([[0,7,9]])
y_t = y.T
y_t
```

```python
# tampilkan ukuran dari vector y_t
y_t.shape
```

###Tensorflow

```python
import tensorflow as tf
```

```python
# buat vector x dengan ukuran 3x1
x_tf = tf.Variable([0,7,9])
x_tf
```

```python
# hitung panjang vector x
tf.rank(x_tf)
```

```python
# tampilkan ukuran vector x
tf.shape(x_tf)
```

```python
# tampilkan tipe data dari vector x
type(x_tf)
```

```python
# akses elemen pertama dari vector x
x_tf[0]
```

```python
# tampilkan tipe data elemen pertama dari vector x
type(x_tf[0])
```

```python
# lakukan transpose pada vector x dan simpan pada variabel x_t
x_tf = tf.Variable([[0,7,9]])
x_tf_t = tf.transpose(x_tf)
x_tf_t
```

```python
# tampilkan ukuran dari vector x_t
tf.shape(x_tf_t)
```

```python
y_tf = tf.Variable([0,7,9])
y_tf
```

```python
# lakukan transpose pada vector y dan simpan pada variabel y_t
y_tf = tf.Variable([[0,7,9]])
y_tf_t = tf.transpose(y_tf)
y_tf_t
```

```python
# tampilkan ukuran dari vector x_t
tf.shape(y_tf_t)
```
