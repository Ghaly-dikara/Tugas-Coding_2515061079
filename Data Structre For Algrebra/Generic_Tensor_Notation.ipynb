###Generic Tensor Notation

```python
import torch
```

```python
# jelaskan tentang perintah ini
images_pt = torch.zeros([32, 28, 28, 3])
images_pt
```

Fungsi torch.zeros() menghasilkan tensor dengan ukuran tertentu yang seluruh nilainya adalah 0.0 (float secara default), sering digunakan untuk inisialisasi data seperti batch gambar sebelum diisi data asli.

Bentuk Tensor (Shape)
Dimensi [32, 28, 28, 3] mewakili:

32: Ukuran batch (jumlah gambar dalam satu kali pemrosesan).

28: Tinggi gambar (height).

28: Lebar gambar (width).

3: Jumlah kanal warna (RGB).
