```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#buat matrix X ukuran 2x2 menggunakan numpy
X = np.array([[1, 2], [3, 4]])
print(X)
```

```python
#buat invers dari matrix X dan beri nama Xinv menggunakan numpy
Xinv = np.linalg.inv(X)
Xinv
```

```python
#kalikan X dengan Xinv menggunakan numpy
np.dot(Xinv, X)
```

```python
import numpy as np
import torch
import tensorflow as tf
```

```python
#selesaikan persamaan diatas menggunakan numPy
X = np.array([[4, 2], [-5, -3]])

y = np.array([4, -7])

hasil_np = np.linalg.solve(X, y)
print(hasil_np)
```

```python
# selesaikan persamaan diatas menggunakan tensorflow
A_tf = tf.constant([
    [4,2],
     [-5,-3]
    ], dtype=tf.float32)
y_tf = tf.constant([[4], [-7]], dtype=tf.float32)

hasil_tf = tf.linalg.solve(A_tf, y_tf)
print(hasil_tf)
```

```python
# selesaikan persamaan diatas menggunakan pytorch
A_pt = torch.tensor ([
    [4,2],
     [-5, -3]
], dtype=torch.float32)
y_pt = torch.tensor([
    [4],
    [-7]
], dtype=torch.float32
)

hasil_pt = torch.linalg.solve(A_pt, y_pt)
print(hasil_pt)
```
