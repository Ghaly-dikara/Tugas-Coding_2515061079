###Matrix Multiplication

```python
# buat matrix A dengan ukuran 2x3 dan vector b berukuran 3x1 menggunakan numpy
import numpy as np
```

```python
A = np.array([[7,9,0],[9,7,0]])
A
```

```python
b = np.array([[0],[7],[9]])
b
```

```python
# tampilkan ukuran matrix A dan vector b
A.shape
```

```python
b.shape
```

```python
# kalikan matrix A dengan vector b
np.dot(A, b)
```

```python
# buat matrix B dengan ukuran 3x3 menggunakan numpy
B = np.array([[7,9,0],[9,7,0],[7,7,7]])
B
```

```python
# kalikan matrix A dengan matrix B
np.dot(A, B)
```

```python
# buktikan apakah perkalian matrix bersifat komutatif? (AB=BA)?

# np.dot(B, A)

# ---------------------------------------------------------------------------
# ValueError                                Traceback (most recent call last)
# /tmp/ipykernel_1722/2086928075.py in <cell line: 0>()
#       1 # buktikan apakah perkalian matrix bersifat komutatif? (AB=BA)?
# ----> 2 np.dot(B, A)
#
# ValueError: shapes (3,3) and (2,3) not aligned: 3 (dim 1) != 2 (dim 0)

''' Jadi perkalian matrix tidak bersifat komutatif dikarenakan perkalian
matrix A (2x3) dan perkalian matrix B (3x3) bisa dilakukan karena memenuhi
bentuk yang memungkinkan untuk dikalikan, tetapi tidak dengan Matrix B (3x3)
dan Matrix A (2x3) karena tidak memenuhi syarat perkalian matrix.
'''
```

###Pytorch

```python
# buat matrix A_pt dengan ukuran 2x3 dan vector b_pt berukuran 3x1 menggunakan pytorch
import torch
```

```python
A_pt = torch.tensor([[7,9,0],[9,7,0]])
A_pt
```

```python
b_pt = torch.tensor([[0],[7],[9]])
b_pt
```

```python
# tampilkan ukuran matrix A dan vector b
A_pt.shape
```

```python
b_pt.shape
```

```python
# kalikan matrix A_pt dengan vector b_pt
torch.matmul(A_pt, b_pt)
```

```python
# buat matrix B_pt dengan ukuran 3x3 menggunakan pytorch
B_pt = torch.tensor([[7,9,0],[9,7,0],[7,7,7]])
B_pt
```

```python
# kalikan matrix A_pt dengan matrix B_pt
torch.matmul(A_pt, B_pt)
```

###TensorFlow

```python
# buat matrix A_tf dengan ukuran 2x3 dan vector b_tf berukuran 3x1 menggunakan tensorflow
import tensorflow as tf
```

```python
A_tf = tf.Variable([[7,9,0],[9,7,0]])
A_tf
```

```python
b_tf = tf.Variable([[0],[7],[9]])
b_tf
```

```python
# tampilkan ukuran matrix A dan vector b
A_tf.shape
```

```python
tf.shape(A_tf)
```

```python
b_tf.shape
```

```python
tf.shape(b_tf)
```

```python
# kalikan matrix A_tf dengan vector b_tf
tf.linalg.matmul(A_tf, b_tf)
```

```python
tf.matmul(A_tf,b_tf)
```

```python
# buat matrix B_tf dengan ukuran 3x3 menggunakan tensorflow
B_tf = tf.Variable([[7,9,0],[9,7,0],[7,7,7]])
B_tf
```

```python
# kalikan matrix A_tf dengan matrix B_tf
tf.linalg.matmul(A_tf, B_tf)
```

```python
tf.matmul(A_tf, B_tf)
```
