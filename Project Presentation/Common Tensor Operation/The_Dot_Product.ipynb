pytorch

```python
import torch #buat vector x_pt dan y_pt dengan ukuran 1x3 menggunakan pytorch
```

```python
x_pt = torch.tensor([19, 8, 2])
x_pt
```

```python
y_pt = torch.tensor([22, 11, 4])
y_pt
```

```python
torch.dot(x_pt, y_pt) #hitung perkalian dot product antara vector x_pt dan vector y_pt
```

```python
torch.dot(torch.tensor([19, 8, 2.]), torch.tensor([22, 11, 4.]))
```

tensorflow

```python
import tensorflow as tf #buat vector x_tf dan y_tf dengan ukuran 1x3 menggunakan tensorflow
```

```python
x_tf = tf.Variable([19, 8, 2])
x_tf
```

```python
y_tf = tf.Variable([22, 11, 4])
y_tf
```

```python
tf.reduce_sum(tf.multiply(x_tf, y_tf)) #hitung perkalian dot product antara vector x_tf dan vector y_tf
```
