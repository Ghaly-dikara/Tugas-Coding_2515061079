```python
import numpy as np
import torch
import tensorflow as tf
```

### Matrix X

```python
# membuat matrix X dengan ukuran 2x3
X = np.array([[25, 2, 3], [5, 26, 7]])
X
```

```python
# membuat matrix X_pt dengan pytorch ukuran 2x3
X_pt = torch.tensor([[25, 2, 3], [5, 26, 7]])
X_pt
```

```python
# membuat matrix X_tf dengan tensorflow ukuran 2x3
X_tf = tf.Variable([[25, 2, 3], [5, 26, 7]])
X_tf
```

Operasi perkalian

```python
# melakukan perkalian matrix X dengan skalar
X*2
```

```python
# melakukan perkalian matrix X_pt pytorch dengan skalar
X_pt*4
```

```python
# melakukan perkalian matrix X_tf tensorflow dengan skalar
X_tf*7
```

operasi penjumlahan

```python
# melakukan perjumlahan matrix X dengan skalar
X + 5
```

```python
# melakukan penjumlahan matrix X_pt pytorch dengan skalar
X_pt + 7
```

```python
# melakukan penjumlahan matrix X_tf tensorflow dengan skalar
X_tf + 11
```

Operasi Pengurangan

```python
# melakukan pengurangan Matrix X dengan skalar
X - 2
```

```python
# melakukan pegurangan Matrix X_pt pytorch dengan skalar
X_pt - 18
```

```python
# melakukan pengurangan Matrix
X_tf - 6
```

### Matrix Y

```python
# membuat Matrix Y dengan ukuran 2x3
Y = np.array([[5, 6, 3], [7, 13, 5]])
Y
```

```python
# membuat matrix Y_pt dengan pytorch ukuran 2x3
Y_pt = torch.tensor([[5, 6, 3], [7, 13, 5]])
Y_pt
```

```python
# membuat matrix Y_tf dengan tensorflow ukuran 2x3
Y_tf = tf.Variable([[5, 6, 3], [7, 13, 5]])
Y_tf
```

```python
# mengalikan matrix X dengan Y
X * Y
```

```python
# mengalikan matrix X_pt dengan Y_pt
X_pt * Y_pt
```

```python
# mengalikan matrix X_tf dengan Y_tf
X_tf * Y_tf
```
