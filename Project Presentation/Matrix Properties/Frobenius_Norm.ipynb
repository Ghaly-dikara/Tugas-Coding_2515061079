###Frobenius Norm

```python
#menggunakan numpy, pytoch dan tensorflow buatlah matrix ukuran 3x3 lalu hitung nilai frobenius norm
import numpy as np
```

```python
X = np.array([[7,9,0],[9,7,0],[7,7,7]])
X
```

```python
X.shape
```

```python
(7**2 + 9**2 + 0**2 + 9**2 + 7**2 + 0**2 + 7**2 + 7**2 + 7**2)**(1/2)
```

```python
(X[0,0]**2 + X[0,1]**2 + X[0,2]**2 + X[1,0]**2 + X[1,1]**2 + X[1,2]**2 + X[2,0]**2 + X[2,1]**2 + X[2,2]**2)**(1/2)
```

```python
np.linalg.norm(X)
```
