```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#buat matrix X ukuran 2x2 menggunakan numpy
X = np.array([[1, 2], [3, 4]])
print(X)
```

```python
#buat invers dari matrix X dan beri nama Xinv menggunakan numpy
Xinv = np.linalg.inv(X)
Xinv
```

```python
#kalikan X dengan Xinv menggunakan numpy
np.dot(Xinv, X)
```

```python
import numpy as np
import torch
import tensorflow as tf
```

```python
X = np.array([[4, 2], [-5, -3]])
X
```

```python
Xinv = np.linalg.inv(X)
Xinv
```

```python
np.dot(Xinv, X)
```

```python
torch.inverse(torch.tensor([[4, 2], [-5, -3.]])) # float type
```

```python
tf.linalg.inv(tf.Variable([[4, 2], [-5, -3.]])) # also float
```
